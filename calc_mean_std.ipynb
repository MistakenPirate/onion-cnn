```python
import torch
from torch.utils.data import DataLoader
from torchvision.datasets import ImageFolder
import torchvision.transforms as transforms
```

```python
# Load dataset as raw tensors (no normalization, just resize + ToTensor)
dataset = ImageFolder(
    root="./Dataset",
    transform=transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])
)

loader = DataLoader(dataset, batch_size=32, shuffle=False)
print(f"Loaded {len(dataset)} images")
```

```python
# Calculate per-channel mean and std across the entire dataset
mean = torch.zeros(3)
std = torch.zeros(3)
n_pixels = 0

for images, _ in loader:
    batch_size = images.size(0)
    images = images.view(batch_size, 3, -1)  # (B, 3, H*W)
    mean += images.sum(dim=[0, 2])
    std += (images ** 2).sum(dim=[0, 2])
    n_pixels += images.shape[0] * images.shape[2]

mean /= n_pixels
std = (std / n_pixels - mean ** 2).sqrt()

print(f"Mean: {mean.tolist()}")
print(f"Std:  {std.tolist()}")
print()
print("Copy this into your config:")
print(f"MEAN = {[round(x, 4) for x in mean.tolist()]}")
print(f"STD = {[round(x, 4) for x in std.tolist()]}")
```
